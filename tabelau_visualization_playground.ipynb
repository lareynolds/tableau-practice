```python
import pandas as pd
pd.options.mode.chained_assignment = None
import random
import numpy as np
```

```python
df = pd.DataFrame(columns=('Category', 'Subcategory', 'Datapoint', 'Zero','One', 'Two', "Three", 'Four', 'Five'))
```

```python
df.head()
```

```python
cat_options = ['Healthy', 'Unhealthy']
sub_options = ['Fruit', 'Grain', 'Meat', 'Plant-based Protein', 'Dairy', 'Fats', 'Vegetable', 'Sweets']
health_options = ['Fruit', 'Grain', 'Vegetable', 'Meat', 'Plant-based Protein', 'Dairy']
unhealth_options = ['Fats', 'Sweets']
```

```python
random_text = [random.choice(cat_options) for _ in range(500)]
```

```python
df ['Category']= random_text
df.head()
```

```python
# prompt: if the Category column contains healthy randomly assign a values from the healthy options

for index, row in df.iterrows():
  if row['Category'] == 'Healthy':
    df.loc[index, 'Subcategory'] = random.choice(health_options)

for index, row in df.iterrows():
  if row['Category'] == 'Unhealthy':
    df.loc[index, 'Subcategory'] = random.choice(unhealth_options)
```

```python
df.head()
```

```python
df.shape
```

```python
fruits = ['Apple', 'Pear', 'Cherry', 'Grapes', 'Orange', 'Banana', 'Peach', 'Plum', 'Strawberries', 'Pineapples', 'Mango', 'Nectarines', 'Melon', 'Papaya', 'Guava', 'Apricots']
grains = ['Wheat', 'Barley', 'Rye', 'Oats', 'Rice', 'Corn', 'Quinoa']
meat = ['Beef', 'Pork', 'Lamb', 'Chicken', 'Turkey', 'Fish', 'Eggs', 'Duck', 'Goose', 'Venison', 'Quail', 'Rabbit']
plant_based_protein = ['Tofu', 'Tempeh', 'Seitan', 'Tempeh', 'Lentils', 'Chickpeas', 'Beans', 'Nuts and Seeds']
dairy = ['Milk', 'Cheese', 'Yogurt', 'Cream']
fats = ['Olive Oil', 'Avacado Oil', 'Grape Seed Oil', 'Butter']
vegetable = ['Corn', 'Potato', 'Tomato', 'Onion', 'Cucumber', 'Lettuce', 'Spinach', 'Broccoli', 'Carrots', 'Cabbage', 'Bell Peppers', 'Zucchini', 'Egg Plant', 'Cauliflower', 'Asparagus', 'Mushrooms', 'Artichokes', 'Peas', 'Squash', 'Beets']
sweets = ['Candy', 'Chocolate', 'Cake', 'Pie', 'Ice Cream', 'Pudding', 'Donuts', 'Cookies']
```

```python
for index, row in df.iterrows():
  if row['Subcategory'] == 'Fruit':
    df.loc[index, 'Datapoint'] = random.choice(fruits)

for index, row in df.iterrows():
  if row['Subcategory'] == 'Grain':
    df.loc[index, 'Datapoint'] = random.choice(grains)

for index, row in df.iterrows():
  if row['Subcategory'] == 'Meat':
    df.loc[index, 'Datapoint'] = random.choice(meat)

for index, row in df.iterrows():
  if row['Subcategory'] == 'Plant-based Protein':
    df.loc[index, 'Datapoint'] = random.choice(plant_based_protein)

for index, row in df.iterrows():
  if row['Subcategory'] == 'Dairy':
    df.loc[index, 'Datapoint'] = random.choice(dairy)

for index, row in df.iterrows():
  if row['Subcategory'] == 'Fats':
    df.loc[index, 'Datapoint'] = random.choice(fats)

for index, row in df.iterrows():
  if row['Subcategory'] == 'Vegetable':
    df.loc[index, 'Datapoint'] = random.choice(vegetable)

for index, row in df.iterrows():
  if row['Subcategory'] == 'Sweets':
    df.loc[index, 'Datapoint'] = random.choice(sweets)
```

```python
df = df.drop_duplicates().reset_index(drop=True)
```

```python
df.shape
```

```python
df.head()
```

```python
def fill_row(row):
    # Generate random integers for each cell such that the sum equals 500
    random_integers = np.random.randint(0, 500, size=6)
    total = random_integers.sum()

    # Scale the random integers to ensure they sum to 500
    if total != 0:
        random_integers = (random_integers / total) * 500

    # Cast the random integers to integers and fill the cells
    row[['Zero', 'One', 'Two', 'Three', 'Four', 'Five']] = random_integers.astype(int)
    return row

# Apply the modified fill_row function to each row in the DataFrame
df = df.apply(fill_row, axis=1)
```

```python
df['Category - Subcategory - Datapoint'] = (df['Category']) + ' - ' + (df['Subcategory']) + ' - ' + (df['Datapoint'])
```
